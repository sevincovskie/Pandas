```python
# 2.Calculate how many days the dataset spans.
spans = houses['date'].max() - houses['date'].min()
spans
```

```python
#  3.Find the date that has the most house sales and all homes sold on that date.
houses['sales_date'] = houses['date'].dt.date
houses['sales_date'].value_counts().head(1)

```

```python
#   4.Create a pie chart showing the years that home sales occurred in.
houses['date'].dt.year.value_counts().plot(kind='pie')
```
