```python
bestsellers = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/bestsellers.csv')
bestsellers
```

```python
# Get the books written by Pete Souza

bestsellers[bestsellers['Author'] == 'Pete Souza']
```

```python
# Get the books below 10 dollars

bestsellers[bestsellers['Price'] < 10]
```

```python
# Get the books with the price between 50 and 60


bestsellers[bestsellers['Price'].between(50,60)]
```

```python
#  Get the books by Kristin Hannah, Andy Weir, or Delia Owens

bestsellers[(bestsellers['Author'] == 'Kristin Hannah') | (bestsellers['Author'] == 'Andy Weir') | (bestsellers['Author'] == 'Delia Owens')]

kristin_hannah = (bestsellers['Author'] == 'Kristin Hannah')
andy_weir = (bestsellers['Author'] == 'Andy Weir')
delia_owens = (bestsellers['Author'] == 'Delia Owens')
books = bestsellers[kristin_hannah | andy_weir | delia_owens]
books

```

```python
# Get the nonfiction books that are rated 4.9
bestsellers[bestsellers['Genre'] == 'Non Fiction']
bestsellers['User Rating'] == 4.9
bestsellers[(bestsellers['Genre'] == 'Non Fiction') & (bestsellers['User Rating'] == 4.9)]
```

```python
# Get the fiction book with the lowest User Rating

bestsellers[bestsellers['Genre'] == 'Fiction'].sort_values('User Rating').head(1)
```

```python
#  Find 2012s top 5 fiction books with the most reviews

fiction = (bestsellers['Genre'] == 'Fiction')
year = (bestsellers['Year'] == 2012)

# bestsellers[fiction & year].sort_values('Reviews', ascending = False).head(5)
bestsellers[fiction & year].nlargest(5, 'Reviews')
```

```python
#  Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

# bestsellers[bestsellers['User Rating'] < 4.5][['Author','User Rating']].value_counts('Author').head().plot(kind='bar')

lower_rated = bestsellers[bestsellers['User Rating'] < 4.5]
lower_rated['Author'].value_counts().head().plot(kind='bar')

```
