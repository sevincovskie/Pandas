```python
# Exercise 2: Colorado Snow Storms report:

# The figure should be 8 by 6
# Colors used: #74b9ff, #7ed6df
# Add Inches text to Y axis


resorts = ['Crested Butte', 'Aspen', 'Telluride', 'Wolf Creek']
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
plt.figure(figsize=(8,6),dpi=150)
plt.title('Colorado Snow Storms')
plt.xlabel('Resort')
plt.ylabel('Inches')
plt.bar(resorts,snow_last_24,color='#74b9ff',label='Last 24 hours')
plt.bar(resorts,snow_next_24,color='#7ed6df',label='Next 24 hours',bottom=snow_last_24)
plt.legend()
```
