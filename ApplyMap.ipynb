```python
def get_age_group(age):
  if age < 2:
    return 'infant'
  elif age < 12:
    return 'child'
  elif age < 18:
    return 'teeenager'
  elif age < 50:
    return 'adult'
  else:
    return 'senior'
```

```python
def compute_family_size (s):
  family_size = s.sibsp + s.parch
  if family_size == 0:
    return f'Solo'
  elif family_size < 5 :
    return f'Average'
  else:
    return f'Large'
```

```python
titanic['family_size'] = titanic.apply(compute_family_size,axis=1)
```
