```python
netflix = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/netflix_titles.csv', sep = "|",index_col=1)
```

```python
# 1.Find the rows with no country
netflix[netflix['country'].isna()]
```

```python
# 2.Find the rows with no country, cast and director listed
netflix[(netflix['country'].isna()) & (netflix['cast'].isna()) & (netflix['director'].isna())]
```

```python
# 3.Drop all rows in the dataframe that have at least one na value (not in place)
netflix.dropna()
```

```python
# 4.Drop all columns in the dataframe that have at least one na value (not in place)
netflix.dropna(axis=1)
```

```python
# 5.Drop all rows in the dataframe that have a missing director or cast (not in place) - research,you need a subset 😜
netflix.dropna(subset=['director','cast'])
```

```python
# 6.Find the rows that are missing a rating. Replace the NA with ‘TV-MA’ using loc

netflix.loc[netflix['rating'].isna(),'rating'] = 'TV-MA'
```

```python
netflix['country'].value_counts()
```

```python
# 7.Fill the missing country values with the most common country. Use fillna instead of loc.
netflix['country'].fillna(netflix['country'])
```

```python
mode_country = netflix['country'].mode()
mode_country
```
