```python
# Import the JoeBidenTweets.csv.
biden = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/JoeBidenTweets.csv',index_col=0)
```

```python
biden
```

```python
# Use the tweet “id” as the index
biden.set_index('id', inplace=True)
```

```python
biden
```

```python
# Drop the “url” column (not in place)
biden.drop(columns='url', axis=1)
```

```python
#  Drop the row with the id of 361388562
biden.drop(labels=361388562, axis=0)
```

```python
#  Add a new column called ‘user’. Set ‘Joe Biden’ for all the rows
biden['user'] = 'Joe Biden'
```

```python
biden
```

```python
#   Add a new column called ‘ratio’. Divide replies by retweets
biden['ratio'] = biden['replies'] / biden['retweets']
```

```python
biden
```

```python
#   Find the top 10 highest ratio tweets
biden.nlargest(n=10, columns='ratio')
```

```python
#   Find the top 10 highest ratio tweets
biden.nlargest(n=10, columns='ratio')
```

```python
biden
```

```python
#    Create a column ‘interaction’ that contains the sum of replies, retweets, quotes and likes
biden['interaction'] = biden['replies'] + biden['retweets'] + biden['quotes'] + biden['likes']
```

```python
biden
```

```python
#    Find the top 8 tweets by number of interactions
biden.nlargest(n=8, columns='interaction')
```
