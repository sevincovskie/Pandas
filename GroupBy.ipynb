```python
# Clean the 'age' and 'fare' columns by replacing '?' with None and converting to float
titanic['age'] = pd.to_numeric(titanic['age'].replace(['?'], [None]), errors='coerce')
titanic['fare'] = pd.to_numeric(titanic['fare'].replace(['?'], [None]), errors='coerce')

# 1. Pclass-a gÃ¶rÉ™ qruplaÅŸdÄ±r: orta saÄŸ qalma (survival rate) vÉ™ sÉ™rniÅŸin sayÄ±
task1 = titanic.groupby('pclass').agg({
    'survived': 'mean',    # Orta saÄŸ qalma faizi (0 vÉ™ 1-lÉ™rin ortalamasÄ±)
    'pclass': 'count' # CÉ™mi sÉ™rniÅŸin sayÄ±
})

# 2. CinsÉ™ (Sex) gÃ¶rÉ™: orta yaÅŸ vÉ™ orta bilet qiymÉ™ti (Fare)
task2 = titanic.groupby('sex')[['age', 'fare']].mean()

# 3. Limana (Embarked) gÃ¶rÉ™: sÉ™rniÅŸin sayÄ± vÉ™ cÉ™mi toplanan pul (Fare sum)
task3 = titanic.groupby('embarked').agg({
    'pclass': 'count',
    'fare': 'sum'
})

# 4. Pclass-a gÃ¶rÉ™ yaÅŸ statistikalarÄ± (min, max, orta, median)
task4 = titanic.groupby('pclass')['age'].agg(['min', 'max', 'mean', 'median'])

# 5. SaÄŸ qalanlara (Survived) gÃ¶rÉ™: orta yaÅŸ, orta bilet, sÉ™rniÅŸin sayÄ±
task5 = titanic.groupby('survived').agg({
    'age': 'mean',
    'fare': 'mean',
    'pclass': 'count'
})
```

```python
# 6. Pclass vÉ™ Sex-É™ gÃ¶rÉ™: saÄŸ qalma faizi vÉ™ orta bilet qiymÉ™ti
task6 = titanic.groupby(['pclass', 'sex']).agg({
    'survived': 'mean',
    'fare': 'mean'
})

# 7. Limana (Embarked) vÉ™ Pclass-a gÃ¶rÉ™: sÉ™rniÅŸin sayÄ± vÉ™ orta yaÅŸ
task7 = titanic.groupby(['embarked', 'pclass']).agg({
    'pclass': 'count',
    'age': 'mean'
})

# 8. SibSp (bacÄ±-qardaÅŸ/hÉ™yat yoldaÅŸÄ±) vÉ™ Parch (valideyn/uÅŸaq) sayÄ±na gÃ¶rÉ™: orta saÄŸ qalma
task8 = titanic.groupby(['sibsp', 'parch'])['survived'].mean()

# 9. Pclass vÉ™ Limana (Embarked)-É™ gÃ¶rÉ™: cÉ™mi bilet pulu vÉ™ orta yaÅŸ
task9 = titanic.groupby(['pclass', 'embarked']).agg({
    'fare': 'sum',
    'age': 'mean'
})

# 10. CinsÉ™ (Sex) vÉ™ SaÄŸ qalanlara (Survived)-É™ gÃ¶rÉ™: orta yaÅŸ vÉ™ orta bilet qiymÉ™ti
task10 = titanic.groupby(['sex', 'survived'])[['age', 'fare']].mean()
```

```python
# 11. Pclass-a gÃ¶rÉ™ É™traflÄ± statistika
# Named Aggregation sintaksisi (sÃ¼tun adlarÄ±nÄ± Ã¶zÃ¼mÃ¼z tÉ™yin edirik)
task11 = titanic.groupby('pclass').agg(
    Total_Passengers=('pclass', 'count'),
    Survivors_Count=('survived', 'sum'), # 1-lÉ™rin cÉ™mi saÄŸ qalanlarÄ±n sayÄ±dÄ±r
    Survival_Rate=('survived', 'mean'),
    Total_Fare=('fare', 'sum')
)

# 12. Limana (Embarked) gÃ¶rÉ™ cinslÉ™r Ã¼zrÉ™ statistika
# Qeyd: Bunu etmÉ™k Ã¼Ã§Ã¼n hÉ™m Embarked, hÉ™m Sex-É™ gÃ¶rÉ™ qruplaÅŸdÄ±rmalÄ±yÄ±q
task12 = titanic.groupby(['embarked', 'sex']).agg({
    'pclass': 'count', # Count by sex
    'age': 'mean',          # Avg age by sex
    'fare': 'mean'          # Avg fare by sex
})

# 13. Pclass vÉ™ Survived-É™ gÃ¶rÉ™: yaÅŸ, bilet vÉ™ ailÉ™ Ã¶lÃ§Ã¼sÃ¼
# ÆvvÉ™lcÉ™ AilÉ™ Ã–lÃ§Ã¼sÃ¼ (FamilySize) sÃ¼tunu yaradaq
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch'] + 1

task13 = titanic.groupby(['pclass', 'survived']).agg({
    'age': 'mean',
    'fare': 'mean',
    'FamilySize': 'mean'
})

# 14. Sex vÉ™ Pclass-a gÃ¶rÉ™ Ã§oxlu statistika
task14 = titanic.groupby(['sex', 'pclass']).agg({
    'pclass': 'count',
    'survived': 'mean',
    'age': ['min', 'max'],
    'fare': 'mean'
})

# 15. Survived vÉ™ Embarked-É™ gÃ¶rÉ™ statistika
task15 = titanic.groupby(['survived', 'embarked']).agg({
    'pclass': 'count',
    'age': 'mean',
    'fare': 'mean',
    'FamilySize': ['mean', 'max'] # AilÉ™ statistikasÄ±
})
```
